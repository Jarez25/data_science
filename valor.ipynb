```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Datos extraídos de la tabla
meses = np.array([1, 2, 3, 4, 5, 6, 7])  # Usamos números para representar los meses
ventas_2023 = [0, 0, 0, 395420.81, 509526.92, 353176.34, 349377.88]
ventas_2024 = [415870.59, 387472.95, 309609.47, 438305.18, 316628.25, 658289.62, 228714.61]

# Creando un DataFrame
df = pd.DataFrame({
    'mes': meses,
    'ventas_2023': ventas_2023,
    'ventas_2024': ventas_2024
})

# Entrenando el modelo de regresión lineal
X = df['mes'].values.reshape(-1, 1)  # Características (meses)
y = df['ventas_2024'].values  # Objetivo (ventas 2024)

model = LinearRegression()
model.fit(X, y)

# Predicción para el mes de agosto (mes 8)
mes_agosto = np.array([[8]])
prediccion_agosto_2024 = model.predict(mes_agosto)

print(f"Predicción de ventas para agosto de 2024: {prediccion_agosto_2024[0]}")

# Visualización de los datos y la predicción
plt.figure(figsize=(10, 6))
plt.plot(meses, ventas_2024, marker='o', label='Ventas 2024')
plt.plot(8, prediccion_agosto_2024, marker='x', color='red', label='Predicción Agosto 2024')

# Añadir títulos y etiquetas
plt.title('Predicción de Ventas para Agosto 2024')
plt.xlabel('Mes')
plt.ylabel('Ventas')
plt.legend()

# Mostrar el gráfico
plt.grid(True)
plt.show()
```
