```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df =  pd.read_csv('planilla_jrodriguez.csv')
```

```python
df
primer = df.iloc[3]
meses = ['ENE', 'FEB', 'MAR', 'ABR', 'MAY', 'JUN', 'JUL', 'AGO', 'SEP', 'OCT', 'NOV', 'DIC']
datos_meses = primer[meses]
```

```python
plt.plot(meses, datos_meses, marker='o', linestyle='-', color='black')
plt.title('Datos de la primera fila por mes')
plt.xlabel('Meses')
plt.ylabel('Valores')
plt.grid(True)
plt.show()
```
