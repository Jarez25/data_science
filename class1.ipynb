```python
import matplotlib.pyplot as plt

anos = [2018, 2019, 2020, 2021, 2022, 2023, 2024]
ventas = [150, 200, 240, 275, 305, 230, 350] 

plt.figure(figsize = (10,5))
plt.plot(anos, ventas, marker='o', linestyle='-', color='black')  # Gráfico de líneas


plt.xlabel('Años')
plt.ylabel('Ventas')
plt.title('Ventas por Año')
plt.grid(True)
plt.show()
```
