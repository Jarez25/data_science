```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#grafico
x = np.linspace(0,10,100)
y = np.sin(x)
#figura y eje
fig, ax = plt.subplots()
ax.plot(x,y)
ax.set_title('Grafico de seno')
ax.set_xlabel('X')
ax.set_ylabel('Seno')
plt.show()
```
